```python
text_path="speech.txt"
def load(filename):
    with open(filename,encoding='utf-8') as f:
        return f.read().lower()

text=load(text_path)
```

```python

def Table(data,k=4):
    T={}
    for i in range(len(data)-k):
        x=data[i:i+k]
        y=data[i+k]
        #print("X",x,"Y",y)
        if T.get(x) is None:
            T[x]={}
            T[x][y]=1
        else:
            if T[x].get(y) is None:
                T[x][y]=1
            else:
                T[x][y]+=1
    return T
    


    
T=Table(text)
```

```python
def prob(T):
    for kx in T.keys():
        s=float(sum(T[kx].values()))
        for k in T[kx].keys():
            T[kx][k]=T[kx][k]/s
    return T


    
```

```python
def train(text,k=5):
    T=Table(text)
    T=prob(T)
    return T
```

```python
model=train(text)
print(model)
```

```python
#sampling
import numpy as np
fruits=["apple","banana","mango"]
prob=[0.8,0,0.2]
for i in range(10):
    print(np.random.choice(fruits,p=prob))
```

```python
def sample(ctx,T,k):
    ctx=ctx[-k:]
    if T.get(ctx) is None:
        return " "
    possible_chars=list(T[ctx].keys())
    possible_values=list(T[ctx].values())
    return np.random.choice(possible_chars,p=possible_values)
```

```python
sample("al",model,4)
```

```python
def generate(initials,k=4,max_len=200):
    sentence=initials
    ctx=initials[-k:]
    for ix in range(max_len):
        next=sample(ctx,model,k)
        sentence+=next
        ctx=sentence[-k:]
    return sentence
```
